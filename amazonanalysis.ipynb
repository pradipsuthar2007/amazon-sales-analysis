# Amazon Sales Analysis Using Pandas

## Project Overview

This project analyzes Amazon sales data to understand sales performance, customer demand, order cancellations, and revenue trends. The dataset contains order information such as product category, order status, fulfillment method, shipping location, order date, and revenue.

The goal of this analysis is to transform raw sales data into meaningful business insights that can help decision-makers improve sales performance and reduce revenue loss.

## Business Questions

The following business questions were investigated:

1. What is the total revenue generated?
2. What percentage of orders are cancelled?
3. Which fulfillment method is used most frequently?
4. Which product category receives the most orders?
5. Which product category generates the highest revenue?
6. Which category has the highest average order value?
7. Which product size is ordered most frequently?
8. Which size generates the highest revenue?
9. Which states generate the most orders?
10. Which states generate the highest revenue?
11. Which cities generate the most orders?
12. Which cities generate the highest revenue?
13. Which month generates the highest revenue?
14. Which month generates the lowest revenue?
15. How does revenue change over time?
16. What is the average order value?
17. What is the revenue contribution of each category?
18. Which category has the highest cancellation rate?
19. How much revenue is lost due to cancelled orders?

## Tools Used

* Python
* Pandas
* Jupyter Notebook

## Expected Outcome

The analysis helps identify top-performing products, high-revenue regions, cancellation patterns, customer purchasing behavior, and potential revenue losses. These insights can support better inventory planning, marketing strategies, and operational decision-making.
